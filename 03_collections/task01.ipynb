```python
grades = [85, 90, 78, 92, 88] # 成績のリスト

average = sum(grades) / 5 # 平均点
max_grade = max(grades) # 最高点
min_grade = min(grades) # 最低点
count_students_above_average = [g for g in grades if g >= average] # 平均以上の点数リスト

print(f"平均点: {average}")
print(f"最高点: {max_grade}")
print(f"最低点: {min_grade}")
print(f"平均以上の学生数: {len(count_students_above_average)}")
```
