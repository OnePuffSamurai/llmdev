```python
# 英単語格納用リスト
words = []

# アルファベットカウント用辞書
alphabet_counts = {'a':0, 'b':0, 'c':0, 'd':0, 'e':0, 'f':0, 'g':0, 'h':0, 'i':0, 'j':0, 'k':0, 'l':0, 'm':0, 'n':0, 'o':0, 'p':0, 'q':0, 'r':0, 's':0, 't':0, 'u':0, 'v':0, 'w':0, 'x':0, 'y':0, 'z':0}

# 空文字が来るまで英単語をリストに格納
while True:
    word = str(input("英単語を入力してください："))

    if len(word) == 0:
        break

    words.append(word)

# 英単語内に登場したアルファベットをカウント
for word in words:
    for letter in word:
        alphabet_counts[letter] += 1

# 英単語リストをABC順に並び替える
sorted_words = sorted(words)

print(f"入力した英単語：{sorted_words}")

for alphabet in alphabet_counts:
    if alphabet_counts[alphabet] > 0:
        print(f"{alphabet}が{alphabet_counts[alphabet]}個ありました")
```
