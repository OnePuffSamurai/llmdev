```python
# ユーザー名のバリデーション関数
def validation(name):
    if len(name) < 5 or len(name) > 15:
        print("エラー: ユーザー名は5文字以上15文字以内で入力してください")
        return False
    elif " " in name:
        print("エラー: ユーザー名に空白を含めることはできません")
        return False
    elif name.isalnum() == False:
        print("エラー: ユーザー名は英数字のみ使用できます")
        return False
    elif 5 <= len(name) <= 15:
        return True
    else:
        return False

user_name = input("ユーザー名を入力してください: ") # ユーザー名入力
checked_name = validation(user_name) # バリデーション関数の呼び出し

if checked_name:
    print("ユーザー名は有効です")
else:
    print("不正なユーザー名です")
```
