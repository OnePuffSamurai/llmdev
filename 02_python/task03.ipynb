```python
principal = int(input("元金を入力してください:")) # 元金
annual_interest_rate = int(input("年利を入力してください:")) # 年利

year = 0 # 年数
final_amount = principal # 最終金額(最初は元金)

while True:
    year += 1

    if year > 10:
        break
    
    final_amount *= (1 + (annual_interest_rate / 100))

print(f"10年後の金額: {final_amount:.2f}")
```
