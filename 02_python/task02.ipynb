```python
price_excl_tax = int(input("税抜価格を入力してください")) # 税抜価格
tax_rate = 0.1 # 消費税率
shipping_cost = 350 # 送料
price_incl_tax = int(price_excl_tax * (1 + tax_rate)) # 税込価格

if price_incl_tax >= 2000:
    print(f"送料は無料です\n送料込税込価格は{price_incl_tax}円です")

else:
    print(f"送料は{shipping_cost}円です\n送料込税込価格は{price_incl_tax + shipping_cost}円です")
```
