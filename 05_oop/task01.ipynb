```python
# 残高管理用クラス
class BankAccount:
    def __init__(self, balance = 1000):
        self.balance = balance

    # 預金メソッド
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
        else:
            print("エラー: 正しい金額を入力してください")

    # 引き出しメソッド
    def withdraw(self, amount):
        if amount < self.balance:
            self.balance -= amount
        elif amount <= 0:
            print("エラー: 残高不足または無効な金額です")
        else:
            print("エラー: 残高不足または無効な金額です")

    # 残高確認メソッド
    def get_balance(self):
        return self.balance

bank = BankAccount()
bank.deposit(int(input("預金額を入力してください: ")))
bank.withdraw(int(input("引き出し額を入力してください: ")))
print(bank.get_balance())
```
